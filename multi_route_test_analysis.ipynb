```python
import re
import pandas as pd
import numpy as np

def filter_and_calculate_average_qor(df):
    # Filter the DataFrame for "hard" and "extreme" tests
       # Filter the DataFrame for "hard" and "extreme" tests
    filtered_df = df[df['Test Name'].str.contains('hard|extreme', case=False)]

    # Calculate average QoR for filtered DataFrame
    qor_average = filtered_df['QoR'].mean()

    return filtered_df, qor_average

def parse_log_data2(log_data):
    # Initialize variables to store parsed data
    test_data = []
    current_test = {}

    # Regular expressions for parsing
    test_name_regex = re.compile(r'Running Tester: (.+)')
    qor_regex = re.compile(r'QoR (.+): (\d+)')
    test_time_regex = re.compile(r'Test time: ([\d.]+) seconds\.')

    # Iterate through each line in the log data
    for line in log_data.split('\n'):
        # Match test name
        test_name_match = test_name_regex.match(line)
        if test_name_match:
            # Start a new test entry
            if current_test:
                test_data.append(current_test)
            current_test = {'Test Name': test_name_match.group(1)}
        # Match QoR
        qor_match = qor_regex.search(line)
        if qor_match:
            current_test['QoR'] = int(qor_match.group(2))
        # Match Test time
        test_time_match = test_time_regex.search(line)
        if test_time_match:
            current_test['Test Time'] = float(test_time_match.group(1))

    # Add the last test entry
    if current_test:
        test_data.append(current_test)

    # Convert parsed data to DataFrame
    df = pd.DataFrame(test_data)
    return df
```
